## Rule-Based Baseline Classifier
We implement a simple keyword-based classifier.

```python
RULES = {
    'spam': ['buy', 'followers', 'free gifts', 'win a prize', 'click here', 'limited time offer', 'http'],
    'self_harm': ['hurt myself', 'might hurt myself', 'not safe', 'hurting myself'],
    'harassment': ['find you', 'ruin your life', 'harassment'],
    'toxic': ['stupid', 'nobody likes you', 'go away'],
}

def rule_predict(text):
    t = text.lower()
    for label, kws in RULES.items():
        if any(kw in t for kw in kws):
            return label
    return 'clean'

df['pred_label_rule'] = df['text'].apply(rule_predict)
df[['text', 'label', 'pred_label_rule']].head()
```

## Evaluation
Compute accuracy and confusion matrix.

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy = accuracy_score(df['label'], df['pred_label_rule'])
conf_matrix = confusion_matrix(df['label'], df['pred_label_rule'])

accuracy, conf_matrix
```
